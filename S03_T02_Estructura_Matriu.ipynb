### Nivell 2 - Exercici 4

Mostreu-me amb exemples de diferents matrius, la regla fonamental de Broadcasting que diu : "les matrius es poden transmetre / broadcast si les seves dimensions coincideixen o si una de les matrius té una mida d'1".

```python
m0 = np.array([2, 4, 6, 8, 0])
m1 = np.array([3, 3, 3, 3, 3])
a = 3

# The same result by multiply 2 arrays with the same shape (one of them full with the same nuber) or t'he array * a single value
print("Multiplicació de matrius 1D amb igual shape o d'una matriu per una variable")
print(m0 * m1)
print(m0 * a)

# broadcast ok
print("\nSuma d'una matriu 2D amb una variable")
print(matriu_2d + a)

# Multiply 1D array * 2D array with the same rows
print("\n2D * 1D amb el mateix nombre de valors a la fila")
print(matriu_2d * m1)

#When two arays have different shape it's possible use "newaxis" to transpose the axis from row to column
# --> https://numpy.org/doc/stable/user/theory.broadcasting.html <--
print("\nUsem NEWAXIS per trasposar la fila a columna, i de dues matrius 1D, n'obtenim una de 2D")

#for that is necessary import the newaxis module
from numpy import newaxis

m2 = np.array([4, 3, 4, 3])
print(m2.shape)
m4 = np.array([8, 7, 2])
print(m4.shape)
new_m = m2[:,newaxis] + m4
new_m2 = m4[:,newaxis] + m2
print(new_m)
print(new_m.shape)
print(new_m2)
print(new_m2.shape)

```

### Nivell 2 - Exercici 6

Mask la matriu anterior, realitzeu un càlcul booleà vectoritzat, agafant cada element i comprovant si es divideix uniformement per quatre.

Això retorna una matriu de mask de la mateixa forma amb els resultats elementals del càlcul.

```python
# array with True or False values depending on the condition
mascara = (matriu_2d % 4 == 0)
print(mascara)
```

```python
#With mask module: import the module
import numpy.ma as ma

#using the matriu_2d array
print(matriu_2d)

# condition: dvisible by 4. 
# Getting the NO divisible values array. The double dash means the divisible values
matriu_no_div4 = ma.masked_where(matriu_2d % 4 == 0, matriu_2d)
print(matriu_no_div4)

# With ~ we get the divisible by 4 values array. The double dash means the non divisible values
matriu_div4 = ma.masked_array(matriu_2d, mask=[~(n % 4 == 0) for n in matriu_2d])
print(matriu_div4)
                                               
```

### Nivell 2 - Exercici 7

Utilitzeu aquesta màscara per indexar a la matriu de números originals. Això fa que la matriu perdi la seva forma original, reduint-la a una dimensió, però encara obteniu les dades que esteu cercant.

```python
# passing mascara array as index to matriu_2d, and get the 1D array with the values divisible by 4.
print(matriu_2d[mascara])

#without using mascara -> using np.where, we get the elements indexes that fullfill the condition.
print(matriu_2d[np.where(matriu_2d % 4 == 0)])
```

```python
%matplotlib inline

#Matplotlib import
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

#loading the image
img = mpimg.imread('../Sprint3/volca.jpg')
#print(img.shape) MxNx3 --> 3 = RGB
#getting few values to know the pixels values
print(img[5, :])

#show the image
imgplot = plt.imshow(img)
```

> Mostreu-me a veure que passa quan eliminem el canal G Verd o B Blau.

```python
 #make a copy from the original
img_no_verd = np.copy(img)

#each sublist is the pixels RGB values => [R=9 G=56 B=23]
# so changing the index 1 to zeros, the green color is eliminated
img_no_verd[:,:, 1] = 0
print(img_no_verd)

#show the image
imgplot = plt.imshow(img_no_verd)
```
